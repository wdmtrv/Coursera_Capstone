# ------------- Part 1--------------

```python
import pandas as pd
import numpy as np
import requests
print("Pandas and NumPy successfully imported")
```

```python
wiki = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
wikipedia_page = requests.get(wiki)

df_raw = pd.read_html(wikipedia_page.content, header=0)[0]
df = df_raw[df_raw.Borough != 'Not assigned']

df.head()
```

Instructed to "ignore Not Assigned cells" - done below

```python
df.loc[df.Neighbourhood == 'Not assigned']
# No Non Assigned neighbourhoods
```

```python
df_group = df.groupby(['Postal Code', 'Borough'])['Neighbourhood'].apply(lambda x: ', '.join(x))
df_group = df_group.reset_index()
df_group.head()
```

```python
df_group.shape
```

# ---------- Part 2 -----------

```python
co_ords = pd.read_csv('https://cocl.us/Geospatial_data')
co_ords.head()
```

```python
df_coords = pd.merge(df,co_ords,on='Postal Code')
df_coords

```

# -------- Part 3 --------

```python
df_toronto = df_coords[df_coords['Borough'].str.contains('Toronto',regex=False)]
df_toronto
```

## Performing the K-means clustering below

```python
from sklearn.cluster import KMeans
k=5
toronto_clustering = df_toronto.drop(['Postal Code','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
```
